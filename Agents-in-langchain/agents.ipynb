```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.tools import tool
import requests
```

```python
from langchain_community.tools import DuckDuckGoSearchRun
search_tool = DuckDuckGoSearchRun()
```

```python
import os 
from dotenv import load_dotenv
load_dotenv()
API_KEY = os.getenv("WEATHERSTACK_API_KEY")
@tool
def get_weather_data(city:str) ->str:
    """
    This Tool fetches the current weather data for a given city
    """
    params={
        'access_key' : 'API_KEY',
        'query' : 'city'
    }
    url='https://api.weatherstack.com/current'

    response = requests.get(url,params)
    return response.json()
```

```python
from langchain_huggingface import ChatHuggingFace,HuggingFaceEndpoint
llm = HuggingFaceEndpoint(
    repo_id = "mistralai/Mistral-7B-Instruct-v0.2",
    task = "text-generation",
    # provider="hf-inference" 

)
llm = ChatHuggingFace(llm=llm)
```

```python
from langchain.agents import create_react_agent,AgentExecutor
# from langsmith import Client
# client=Client()

prompt = AgentExecutor.pull('hwchase17//react')
# system_prompt_str = prompt.messages[0].prompt.template

```

```python
agent =create_react_agent(
    llm=llm,
    tools=[search_tool,get_weather_data],
    prompt=prompt
)
```

```python
agent_executor = AgentExecutor(
    agent=agent,
    tools=[search_tool,get_weather_data],
    verbose=True
)
```

```python
response = agent_executor.invoke({"input":"Find the capital of nepal , then find it\'s current weather condition"})
print(response)
```
